```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\Adhil E\\Downloads\\Dataset_II.csv")
```

```python
df.isnull().sum()
```

```python
df['Saving accounts'] = df['Saving accounts'].fillna(df['Saving accounts'].mode()[0])
df['Checking account'] = df['Checking account'].fillna(df['Checking account'].mode()[0])
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df=pd.get_dummies(data=df,columns=['Sex','Housing','Saving accounts','Checking account','Purpose'],drop_first=True)
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
distortions=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df)
    distortions.append(kmeans.inertia_)
```

```python
x=np.arange(1,10)
```

```python
plt.plot(x,distortions)
plt.xlabel("Value of K")
plt.ylabel("Inertia")
plt.title("K vs Inertia")
```

```python
#SO OPTIMUM K IS 2
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
ypred = kmeans.fit_predict(df)
plt.scatter(df.iloc[:,2], df.iloc[:,0],c=ypred)
plt.xlabel("Credit Amount")
plt.ylabel("Age")
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 0], s=30, c='red',label='centre')
plt.title("K Means Clustering")
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
gm=GaussianMixture(n_components=2)
```

```python
gm.fit(df)
```

```python
ypredgm=gm.predict(df)
```

```python
plt.scatter(df.iloc[:,2], df.iloc[:,0],c=ypredgm)
plt.xlabel("Credit Amount")
plt.ylabel("Age")
plt.title("Gaussian Mixture Clustering")
plt.show()
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(df)
data_scaled = pd.DataFrame(df, columns=df.columns)
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 10))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y=80000, color='r', linestyle='--')
```

```python
#SO NUMBER OF CLUSTERS IS 2
```

```python
from sklearn.cluster import AgglomerativeClustering
agc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
ypredagc=agc.fit_predict(data_scaled)
```

```python
plt.scatter(df.iloc[:,2], df.iloc[:,0],c=ypredagc)
plt.xlabel("Credit Amount")
plt.ylabel("Age")
plt.title("Hierarchical Clustering")
plt.show()
```

```python
dfdbscan=df[["Credit amount","Age"]]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
dfdbscan=pd.DataFrame(sc.fit_transform(dfdbscan),columns=dfdbscan.columns)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = .4, min_samples = 20).fit(dfdbscan)
```

```python
ypreddb=dbscan.labels_
```

```python
plt.scatter(df.iloc[:,2], df.iloc[:,0],c=ypreddb)
plt.xlabel("Credit Amount")
plt.ylabel("Age")
plt.title("DBSCAN")
plt.legend()
plt.show()
```
