```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\Adhil E\\Downloads\\Dataset_II.csv")
```

```python
df.isnull().sum()
```

```python
df['Saving accounts'] = df['Saving accounts'].fillna(df['Saving accounts'].mode()[0])
df['Checking account'] = df['Checking account'].fillna(df['Checking account'].mode()[0])
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df=pd.get_dummies(data=df,columns=['Sex','Housing','Saving accounts','Checking account','Purpose'],drop_first=True)
```

```python
ytrain=df.iloc[:,5]
```

```python
df=df.drop(columns=['Housing_own'])
```

```python
xtrain=df
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
annmodel2=Sequential()
annmodel2.add(Dense(4,input_dim=18,activation='relu'))
annmodel2.add(Dense(6,activation='relu'))
annmodel2.add(Dense(1,activation='sigmoid'))
annmodel2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
annmodel2.fit(xtrain,ytrain,epochs=25,batch_size=25)
```

```python
annmodel2.predict_classes(xtrain)
```
