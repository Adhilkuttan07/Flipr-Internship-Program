```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("C:\\Users\\Adhil E\\Downloads\\Train_Dataset-01.csv")
test=pd.read_csv("C:\\Users\\Adhil E\\Downloads\\Test_Dataset-01.csv")
```

```python
train=train.fillna(train.mean())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
train['diagnosis']=le.fit_transform(train['diagnosis'])
```

```python
train=train.drop(columns=['id'])
test=test.drop(columns=['id'])
```

```python
ytrain=train.iloc[:,0]
xtrain=train.iloc[:,1:]
```

```python
xtest=test
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
annmodel=Sequential()
annmodel.add(Dense(4,input_dim=30,activation='relu'))
annmodel.add(Dense(5,activation='relu'))
annmodel.add(Dense(1,activation='sigmoid'))
annmodel.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
annmodel.fit(xtrain,ytrain,epochs=100,batch_size=10)
```

```python
ypred=annmodel.predict_classes(xtest)
```
